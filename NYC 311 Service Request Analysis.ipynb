```python
# Convert 'Created Date' and 'Closed Date' to datetime format
df['Created Date'] = pd.to_datetime(df['Created Date'], errors='coerce')
df['Closed Date'] = pd.to_datetime(df['Closed Date'], errors='coerce')
```

```python
# Create new column for request closing time (in days)
df['Request_Closing_Time'] = (df['Closed Date'] - df['Created Date']).dt.total_seconds() / (60 * 60 * 24)
df[['Created Date', 'Closed Date', 'Request_Closing_Time']]
```

```python
# Drop irrelevant columns

columns_to_drop = [
    'Agency Name', 'Incident Address', 'Street Name', 'Cross Street 1', 'Cross Street 2',
    'Intersection Street 1', 'Intersection Street 2', 'Address Type', 'Park Facility Name',
    'Park Borough', 'School Name', 'School Number', 'School Region', 'School Code',
    'School Phone Number', 'School Address', 'School City', 'School State', 'School Zip',
    'School Not Found', 'School or Citywide Complaint', 'Vehicle Type', 'Taxi Company Borough',
    'Taxi Pick Up Location', 'Bridge Highway Name', 'Bridge Highway Direction', 'Road Ramp',
    'Bridge Highway Segment', 'Garage Lot Name', 'Ferry Direction', 'Ferry Terminal Name',
    'Landmark', 'X Coordinate (State Plane)', 'Y Coordinate (State Plane)', 'Due Date',
    'Resolution Action Updated Date', 'Community Board', 'Facility Type', 'Location'
]
```

```python
# Drop columns only if they exist in the DataFrame
df.drop(columns=[col for col in columns_to_drop if col in df.columns], inplace=True)
print(df.columns)
```

```python
# Remove rows with missing values
df.dropna(inplace=True)
df.shape
```

```python
# Print unique values for all columns
for column in df.columns:
    print(f"\nColumn: {column}")
    print(df[column].unique())
```

```python
# Create a copy for numerical analysis
df_copy = df.copy()
```

```python
# Convert 'Request_Closing_Time' to total seconds (optional if you already have in days)
df_copy['Request_Closing_Time'] = df_copy['Request_Closing_Time'] * 24 * 60 * 60  # Convert back to seconds
```

```python
# Select only numerical columns
numeric_df = df_copy.select_dtypes(include='number')
```

```python
# Compute summary statistics
summary_stats = pd.DataFrame({
    'Sum': numeric_df.sum(),
    'Mean': numeric_df.mean(),
    'Standard Deviation': numeric_df.std(),
    'Skewness': numeric_df.skew(),
    'Kurtosis': numeric_df.kurt()
})
print("\nSummary Statistics:")
print(summary_stats)
```

```python
# Correlation matrix heatmap
correlation_matrix = numeric_df.corr()

plt.figure(figsize=(10, 8))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar(label='Correlation Coefficient')
plt.xticks(ticks=range(len(correlation_matrix.columns)), labels=correlation_matrix.columns, rotation=90)
plt.yticks(ticks=range(len(correlation_matrix.index)), labels=correlation_matrix.index)
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```

```python
#Insight 3: Distribution of Request Closing Time

closing_hours = df['Request_Closing_Time'] * 24  # Convert days to hours
closing_hours = closing_hours.dropna()

# Create log-spaced bins
bins = np.logspace(np.log10(0.1), np.log10(closing_hours.max()), 60)

# Plot histogram
plt.figure(figsize=(16, 6))
plt.hist(closing_hours, bins=bins, color='mediumseagreen', edgecolor='black')
plt.xscale('log')

# Set readable x-ticks 
ticks = [0.1, 0.5, 1, 2, 5, 10, 20, 50, 100, 200, 500, 1000]
tick_labels = [f"{t}h" for t in ticks]
plt.xticks(ticks, tick_labels)

# Axis labels and styling
plt.title("Distribution of Request Closing Time (Hours)", fontsize=16)
plt.xlabel("Request Closing Time (in Hours)", fontsize=14)
plt.ylabel("Number of Requests", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout(pad=2)
plt.show()
```

```python
# Insight 4: Monthly Trend of Complaints

df['Month'] = df['Created Date'].dt.to_period('M')
monthly_counts = df['Month'].value_counts().sort_index()

plt.figure(figsize=(14,6))
plt.plot(monthly_counts.index.astype(str), monthly_counts.values, color='darkblue', marker='o')
plt.title("Monthly Complaint Volume Trend")
plt.xlabel("Month")
plt.ylabel("Number of Complaints")
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Avg Request Closing Time by Complaint Type and Borough

df['Closing_Hours'] = df['Request_Closing_Time'] * 24

grouped = df.groupby(['Complaint Type', 'Borough'])['Closing_Hours'].mean().unstack()

# Filter top 10 complaint types
top5_complaints = df['Complaint Type'].value_counts().nlargest(10).index
filtered_grouped = grouped.loc[top5_complaints]

filtered_grouped.plot(kind='bar', figsize=(14,7), colormap='tab10')
plt.title("Avg Request Closing Time by Complaint Type and Borough")
plt.ylabel("Avg Closing Time (Hours)")
plt.xlabel("Complaint Type")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import f_oneway, chi2_contingency
```

```python
# Test 1: Are average response times similar across complaint types?

# Filter top 5 complaint types to keep the test manageable and meaningful
top_complaints = df['Complaint Type'].value_counts().nlargest(5).index
filtered_df = df[df['Complaint Type'].isin(top_complaints)]

# Group data by complaint type and prepare samples for ANOVA
samples = [group['Closing_Hours'].dropna() for _, group in filtered_df.groupby('Complaint Type')]

# Perform one-way ANOVA
f_statistic, p_value = f_oneway(*samples)

# Output the result
print("ANOVA Test - Average Request Closing Time across Complaint Types")
print("F-statistic:", f_statistic)
print("p-value:", p_value)

# Interpretation
if p_value < 0.05:
    print("Conclusion: Reject the null hypothesis — At least one complaint type has a different average closing time.")
else:
    print("Conclusion: Fail to reject the null hypothesis — No significant difference in average closing times.")
```

```python
# Test 2: Are Complaint Type and Borough Related?

# Generate the contingency table using full data
contingency_table_full = pd.crosstab(df['Complaint Type'], df['Borough'])

# Perform the Chi-Square test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table_full)

# Print the results
print("Chi-square Statistic:", chi2_stat)
print("Degrees of Freedom:", dof)
print("p-value:", p_value)

# Interpretation
if p_value < 0.05:
    print("Conclusion: Reject the null hypothesis")
else:
    print("Conclusion: Fail to reject the null hypothesis")
```
